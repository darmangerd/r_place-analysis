```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
shape = (1000, 1000)
d = np.indices(shape).transpose((1,2,0)).reshape((np.multiply(*shape), 2))

df = pd.DataFrame(d, columns=["x", "y"])
df
```

```python
plt.figure(figsize=(90, 90), dpi=300) 

fig = sns.scatterplot(data=df, x="x", y="y", size=0.1)
fig.set(title="")
fig.set(xticklabels=[])
fig.set(xlabel=None)
fig.set(yticklabels=[])
fig.set(ylabel=None)
```

```python
shape = (38, 10)
d = np.indices(shape).transpose((1,2,0)).reshape((np.multiply(*shape), 2))

df = pd.DataFrame(d, columns=["x", "y"])
df
```

```python
plt.figure(figsize=(5, 2), dpi=500) 

fig = sns.scatterplot(data=df, x="x", y="y", marker="s")
fig.set(title="")
fig.set(xticklabels=[])
fig.set(xlabel=None)
fig.set(yticklabels=[])
fig.set(ylabel=None)
```
