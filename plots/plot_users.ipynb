# Imports

```python
import pandas as pd
import plotly.express as px
import os
```

# Load datas

```python
src_path = os.path.join("/", "Volumes", "owen", "datas", "rplace")
```

```python
df_user_clicks_count = pd.read_csv(os.path.join(src_path, "user_clicks_count.csv"))
```

```python
df_user_per_clicks = pd.DataFrame(df_user_clicks_count.value_counts(["index"]).sort_index(), columns=["count"])
df_user_per_clicks.reset_index(inplace=True)
df_user_per_clicks["index"] = df_user_per_clicks["index"] - 1
```

# Most active users

```python
args = dict(
    data_frame=df_user_per_clicks,
    x="index",
    y="count",
    log_y=False,
    title="Nombre d'utilisateurs ayant cliqué un certain nombre de fois sur le canvas de l'évenement r/place de 2022",
    labels={"index": "Nombre de clicks", "count": "Nombre d'utilisateur"},
    height=600
)

fig = px.bar(**args)
fig.write_html(f"html/users_clicks.html")
fig.show()

args["log_y"] = True
args["title"] = f"{args['title']} (échelle logarithme)"
fig = px.bar(**args)
fig.write_html(f"html/users_clicks_log.html")
fig.show()
```
