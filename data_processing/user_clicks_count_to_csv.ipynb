```python
import pandas as pd
import os
from glob import glob
from typing import List
```

# Files

```python
src_path = os.path.join("/", "Volumes", "owen", "datas", "rplace")
src_glob = os.path.join(src_path, "2022_place_canvas_history-*.csv")
dest_path = os.path.join(src_path)
```

```python
files = glob(src_glob)
files
```

# Data processing

```python
def count_user_clicks(filename: str):
    df_place = pd.read_csv(filename)
    df_place.reset_index(inplace=True)
    df_user_click_count = df_place.groupby("user_id").count()
    return df_user_click_count[["index"]].rename({"index": "count"}, axis=1)
```

```python
count_user_clicks(filename=files[0])
```

```python
def merge_count_user_clicks(files: List[str]):
    df: pd.DataFrame = count_user_clicks(files[0])

    for i in range(1, len(files)):
        print(f"Adding {files[i]} ({i / len(files) * 100}%)")
        df_to_add = count_user_clicks(files[i])
        df = df.add(df_to_add, fill_value=1)
    
    return df.sort_values("index", ascending=False)
```

```python
df = merge_count_user_clicks(files[:])
df
```

```python
df.to_csv(os.path.join(dest_path, "user_clicks_count.csv"))
```
